```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the metrics, 20 m band sharpening
models = [
    "Bicubic",
    "DSen2",
    "Zhang et al.",
    "Sen2-RDSR",
    "David J"
]

metrics = {
    "RMSE ()": [125.68, 35.85, 34.99, 34.38, 29.00],
    "SRE ()": [26.44, 35.94, 36.19, 36.38, 24.53],
    "SAM ()": [1.21, 0.78, 0.75, 0.75, 3.59],
    "PSNR ()": [45.82, 55.54, 55.77, 55.94, 32.08],
    "SSIM ()": [0.82, 0.93, 0.93, 0.93, 0.98],
    "ERGAS ()": [3.33, 1.07, 1.03, 1.02, 3.65],
}

# Define whether lower or higher is better for each metric
metric_goal = {
    "RMSE ()": "min",
    "SRE ()": "max",
    "SAM ()": "min",
    "PSNR ()": "max",
    "SSIM ()": "max",
    "ERGAS ()": "min",
}

# Define rosa color for David J
rosa_color = "#003366"  # Dark blue

n_metrics = len(metrics)
fig, axes = plt.subplots(1, n_metrics, figsize=(4 * n_metrics, 6))

if n_metrics == 1:
    axes = [axes]  # Ensure axes is iterable even for a single metric

for ax, (metric_name, values) in zip(axes, metrics.items()):
    # Determine the winning value(s)
    comparator = np.argmin if metric_goal[metric_name] == "min" else np.argmax
    best_value = (min if metric_goal[metric_name] == "min" else max)(values)

    # In case of ties, mark all winners
    winners = [i for i, v in enumerate(values) if abs(v - best_value) < 1e-9]

    # Assign colors: rosa for David J, blue for other winners, gray for others
    colors = []
    for i in range(len(values)):
        if models[i] == "David J":
            colors.append(rosa_color)
        elif i in winners:
            colors.append("#0072B2")
        else:
            colors.append("#D3D3D3")

    # Plot bar chart
    bars = ax.bar(models, values, color=colors)
    # Make the text above each column (the value labels) bigger
    ax.set_title(metric_name, fontsize=24, pad=15, fontweight="bold")
    ax.set_xticklabels(models, rotation=45, ha="right")

    # Annotate bars with numeric values
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax.annotate(f"{value:.2f}",
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 0),  # 3 points vertical offset
                    textcoords="offset points",
                    ha="center", va="bottom", fontsize=13)  # Increased fontsize here

    # Add grid for readability
    ax.yaxis.grid(True, linestyle="--", alpha=0.5)

fig.suptitle("Quantitative Evaluation Metrics for 20 m Band Sharpening", fontsize=16, y=1.02, fontweight="bold")
fig.tight_layout()
plt.show() 
```
