```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the metrics, 20 m band sharpening
models = [
    "Bicubic",
    "DSen2",
    "Zhang et al.",
    "Sen2-RDSR",
]

metrics = {
    "RMSE (↓)": [125.68, 35.85, 34.99, 34.38],
    "SRE (↑)": [26.44, 35.94, 36.19, 36.38],
    "SAM (↓)": [1.21, 0.78, 0.75, 0.75],
    "PSNR (↑)": [45.82, 55.54, 55.77, 55.94],
    "SSIM (↑)": [0.82, 0.93, 0.93, 0.93],
    "ERGAS (↓)": [3.33, 1.07, 1.03, 1.02],
}

# Define whether lower or higher is better for each metric
metric_goal = {
    "RMSE (↓)": "min",
    "SRE (↑)": "max",
    "SAM (↓)": "min",
    "PSNR (↑)": "max",
    "SSIM (↑)": "max",
    "ERGAS (↓)": "min",
}

n_metrics = len(metrics)
fig, axes = plt.subplots(1, n_metrics, figsize=(4 * n_metrics, 6))

if n_metrics == 1:
    axes = [axes]  # Ensure axes is iterable even for a single metric

for ax, (metric_name, values) in zip(axes, metrics.items()):
    # Determine the winning value(s)
    comparator = np.argmin if metric_goal[metric_name] == "min" else np.argmax
    best_value = (min if metric_goal[metric_name] == "min" else max)(values)

    # In case of ties, mark all winners
    winners = [i for i, v in enumerate(values) if abs(v - best_value) < 1e-9]

    # Assign colors: blue for winners, gray for others
    colors = ["#0072B2" if i in winners else "#D3D3D3" for i in range(len(values))]

    # Plot bar chart
    bars = ax.bar(models, values, color=colors)
    ax.set_title(metric_name, fontsize=16, pad=15, fontweight="bold")
    ax.set_xticklabels(models, rotation=45, ha="right")

    # Annotate bars with numeric values
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax.annotate(f"{value:.2f}",
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha="center", va="bottom", fontsize=8)

    # Add grid for readability
    ax.yaxis.grid(True, linestyle="--", alpha=0.5)

fig.suptitle("Quantitative Evaluation Metrics for 20 m Band Sharpening", fontsize=16, y=1.02, fontweight="bold")
fig.tight_layout()
plt.show() 
```
